```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('index.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.set_index('date')
```

```python
df.head()
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
df['day'] = df['datetime'].dt.day
df['day_name'] = df['datetime'].dt.day_name()
df['hour'] = df['datetime'].dt.hour

df.head()
```

```python
coffee_names = df['coffee_name'].unique()
coffee_names
```

```python
Sales_month = df.sort_values(by='day').groupby('day').size()

plt.figure(figsize=(12,8))
plt.title('Coffe Month Sales')
sns.barplot(x=Sales_month.index,y= Sales_month,palette='crest')
plt.show()
```

```python
for coffee_name in coffee_names:
    sales = df[df['coffee_name'] == coffee_name].groupby('day').size()
    plt.figure(figsize=(12,8))
    plt.title(f'{coffee_name} Sales Across the Month')
    sns.barplot(x=sales.index, y=sales, palette='crest')
    plt.show()
```

```python
# Sales on Week

Week_Sales = df.groupby('day_name').size()
Week_Sales= Week_Sales.reindex(index= ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
plt.figure(figsize=(8,6))
plt.title("Sales On Week")
sns.barplot(x=Week_Sales.index,y=Week_Sales,palette='flare')
plt.xlabel('Day')
plt.show()

```

```python
for coffee_name in coffee_names:
    sales = df[df['coffee_name'] == coffee_name].groupby('day_name').size()
    sales = sales.reindex(index=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
    plt.figure(figsize=(8,6))
    plt.title(f'{coffee_name} Sales Across the Week_Days')
    sns.barplot(x=sales.index, y=sales, palette='crest')
    plt.show()
```

```python
X = df['cash_type'].value_counts()
labels = X.index
val = X.values
val,labels,X
```

```python
fig = plt.pie(val,labels=X.index)
plt.title('Pie Chart (Card vs Cash)')
```

```python
import matplotlib.pyplot as plt

# Assuming 'val' is a list or array containing the values for the pie chart
# and 'X' is a Series/DataFrame with the 'cash_type' labels

# Create the pie chart
fig, ax = plt.subplots(figsize=(6, 6))
ax.pie(val, labels=X.index, autopct='%1.1f%%')

# Add title and other styling
plt.title('Pie Chart (Card vs Cash)')
ax.axis('equal')  # Equal aspect ratio ensures that pie is circular.
plt.show()
```
